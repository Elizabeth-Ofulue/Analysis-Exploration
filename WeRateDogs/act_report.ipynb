# Data Visualisation of `WeRateDogs` Twitter Archive

### Introduction 

> Using Python and its libraries, data will be gathered from a variety of sources and formats for the tweet archive of Twitter user `@dog_rates`, also known as `WeRateDogs`. After gathering the data quality and tidiness is assessed and cleaned.

> This report communicates the insights and displays the visualisations produced from the data wrangling process of the `WeRateDogs` Twitter dataset. After wrangling and cleaning, the dataset was analysed to answer some questions including:

> * Top and least dogs per dog breed.
> * Most populated dog stage.
> * Dogs with the highest average ratings and likes.
> * Dog stage with the highest likes and ratings.
> * Do dogs with high ratings have higher likes.

### Dogs per Breed

> The first bar chart below communicates the top 10 most predicted dog breed from the `WeRateDogs` Twitter dataset. It can be seen that the most predicted dog breed is the `Golden Retriever` with almost 160 dogs predicted as such.

### Dogs per Dog Stage
> Grouping by dog stage to obtain the number of dogs present in each stage, it can be seen that `pupper` is the most populated dog stage with more than 60% of dogs in this stage. This is followed by `doggo` and `puppo` stages.  From the graph it can also be observed that `doggo, pupper` and `doggo, floofer` stages contain the same percentage of dogs.

### Average Likes and Ratings Per Dog Breed

> Grouping the dataset by average rating per dog breed, it can be observed that although `Bouvier Des Flandres` breed has the highest average ratings, it doesn't have the highest likes (favorite count).

> Grouping the data by average likes (favorite count) per dog breed, it can be observed that `Bedington Terrier` has the highest average favorite count.

### Average Likes and Ratings Per Dog Stage

> Grouping the dataset by average rating per dog stage, it can be observed that `doggo, puppo` stage has the highest percentage of average rating, followed by `puppo` and `floofer` stages which share the same percentage of rating. `doggo, floofer` and `doggo, pupper` also share the same percentage of average rating. 

> Grouping the data by average likes (favorite count) per dog stage, it can be observed that `doggo, puppo` stage has the highest percentage of average likes, followed by `puppo` and `doggo` stages.

### Likes and Ratings
> A line chart is used in this case to show the trend between various ratings and the average number of likes(favorite_count) per tweet. This is used to understand if dogs with high ratings have a higher number of likes and vice-versa. 

> It can be observed that some dogs with high ratings tend to have a high number of likes, the number of likes tend to diminish as the ratings decreaserease below `0.6`. For dogs with ratings from `0.9` and above, an increase is observed in the number of likes received.

### Conclusion

> From the charts above which are used to show the relationship between variables, it can be seen that tweets with high likes (favorite count) show high ratings for per tweet.
